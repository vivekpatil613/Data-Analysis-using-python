Data Cleaning

```python
#Deleting unncecessary Columns
data = df.drop(['Web_Link', 'Phone_number','Rest_timming','4_star_review_percentage','3_star_review_percentage','2_star_review_percentage'], axis=1)
```

```python
# duplicate values control and delete duplicates with columns "	Restaurant_Name",'Locality'
data = data.drop_duplicates(subset=["Restaurant_Name",'Locality'], keep='last')
```

```python
data.shape
```

```python
#Removing unecessary values from columns
data['Number of votes'].replace(r'votes', '', regex=True, inplace=True)
```

```python
data['Charges_for_two'].replace(r'₹', '', regex=True, inplace=True)
data['Charges_for_two'].replace(r',', '', regex=True, inplace=True)
```

```python
data['5_star_review_percentage'].replace(r'%', '', regex=True, inplace=True)
data['1_star_review_percentage'].replace(r'%', '', regex=True, inplace=True)
```

Visualization

```python
def count_plot(data ,x ,y): 
    plt.figure(figsize=(x,y))
    ax = sns.countplot(x=data, palette= "Spectral" , edgecolor='white', linewidth=1, order=data.value_counts().iloc[0:10].index)
    plt.xticks(rotation=90)
```

```python
count_plot(data['Locality'],10,6)
```

```python
count_plot(df['Desserts and Bakes'],3, 5)
```

```python
count_plot(df['Home Delivery'],3, 5)
```

```python
sns.countplot(data['payment_modes'])
sns.countplot(data['payment_modes']).set_xticklabels(sns.countplot(data['payment_modes']).get_xticklabels(), rotation=90, ha='right')
fig = plt.gcf()
fig.set_size_inches(15,8)
plt.title('Type of Payment methods')
```

```python
from wordcloud import WordCloud 

word_df = df['Cuisines']
word_df = word_df.dropna(how="any", axis=0)

word_df= word_df.str.replace(r',', '', regex=True)
text =' '.join(word_df)

wordcloud= WordCloud(font_path='/content/CabinSketch-Bold.ttf',
                        width=2400, height=1200,
                        background_color="white", max_words=100, colormap="Accent").generate(text)
# Generate plot
plt.figure( figsize=(10,10), facecolor='k')
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
